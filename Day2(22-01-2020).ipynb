### Without Arg Without Return
`

    Statement

```python
# Without Arg Without Return

def woawor():
    print(2+3)
woawor()
```

```python
# with arg without return
def wawor(a,b):
    print(a+b)
wawor(4,5)
    
```

```python
# without arg with return
def wawor():
    return 5+4
wawor()


```

```python
# with Arg with Return
def sub(a,b):
    return a-b
sub(10,5)
```

```python
def sum1(a,b):
    return a+b
def sub(a,b):
    return a-b
def Mul(a,b):
    return a*b
def Div(a,b):
    return a/b
```

```python
print(sum1(3,5))
print(sub(5,9))
print(Mul(4,99))
print(Div(999,3))
```

```python
a= int(input())
c=0
for i in range(1,a+1):
    if a%i==0:
        c +=1
if c == 2:
    print("It is Prime")
else:
    print("Not Prime")
    
```

```python
# 6 --> 1,2,3
a=int(input("Enter Number"))
s=0
for i in range(1,a):
    if a%i==0:
        s += i
if s==a:
    print("Perfect Number")
else:
    print('Not Perfect Number')
```
